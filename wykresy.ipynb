```python
from ipynb.fs.full.mnk_numeryczna import *
from ipynb.fs.full.mnk_analityczna import *
import numpy as np
import scipy as sp
import math
import matplotlib.pyplot as plt
from scipy import optimize
from pandas import DataFrame , read_excel
from pylab import 
from random import *
```

```python
#dane
pomiary = read_excel('pomiary1.xlsx')
stopa=list(pomiary['stopa'])
wydatki=list(pomiary['wydatki'])
wydatki,stopa=clean_nans(wydatki,stopa)
wydatki= np.array(wydatki).reshape(1,len(wydatki))+randint(1,100)
stopa= np.array(stopa).reshape(1,len(stopa))+randint(1,5)
```

```python
#analityczna 
an=analityczna(stopa,wydatki)
a1=np.array(an[0])
a2=np.array(an[1])
```

```python
#numeryczna 
st_1 = np.array([0,0])
st_2 = np.array([0,0,0])
st_3 = np.array([0,0,0,0])
f1 = lambda a:(mnk1(a, wydatki, stopa))
f2 = lambda a:(mnk2(a, wydatki, stopa))
f3 = lambda a:(mnk3(a, wydatki, stopa))
f4 = lambda a:(mnk4(a, wydatki, stopa))
f5 = lambda a:(mnk5(a, wydatki, stopa))
f5 = lambda a:(mnk5(a, wydatki, stopa))
f6 = lambda a:(mnk_sin(a, wydatki, stopa))
opt1=optymalizuj(f1,st_1)
opt2=optymalizuj(f2,st_2)
opt3=optymalizuj(f3,st_1)
opt4=optymalizuj(f4,st_2)
opt5=optymalizuj(f5,st_3)
opt6=optymalizuj(f6,st_1)
```

```python
#osie
mini=int(np.min(wydatki))
maxi=int(np.max(wydatki))
os_x = np.linspace(mini,maxi,num=1000)
os_y_ep1 = os_x*opt1[0]+opt1[1]
os_y_ep2 = pow(os_x,2)*opt2[0]+opt2[1]*os_x+opt2[2]
os_y_abs1 = os_x*opt3[0]+opt3[1]
os_y_abs2 = pow(os_x,2)*opt4[0]+opt4[1]*os_x+opt4[2]
os_y_abs3 = pow(os_x,3)*opt5[0]+opt5[1]*pow(os_x,2)+opt5[2]*os_x+opt5[3]
os_y_abs4 = np.sin(opt6[0]*os_x)+opt6[1]
os_y_an2 = a1[0]*pow(os_x,2)+a1[1]*os_x+a1[2]
os_y_an1 = a2[0]*os_x+a2[1]
```

```python
#wykres - proste
plt.figure(num=None, figsize=(20, 10), dpi=100)
plt.plot(os_x,os_y_ep1,'b')
plt.plot(os_x,os_y_abs1,'violet')
plt.plot(os_x,os_y_an1,'y')
plt.plot(wydatki,stopa,'r*')
plt.xlabel('Wydatki')
plt.ylabel('Stopa')
plt.gca().legend(('e^2_prosta','abs_prosta','an_prosta','data'))
plt.show()
```

```python
#wykres - mix
plt.figure(num=None, figsize=(20, 10), dpi=200, facecolor='w', edgecolor='k')
plt.plot(os_x,os_y_ep2,'y')
plt.plot(os_x,os_y_abs2,'g')
plt.plot(os_x,os_y_an2,'black')
plt.plot(os_x,os_y_abs4,'magenta')
plt.plot(wydatki,stopa,'r*')
plt.xlabel('Wydatki')
plt.ylabel('Stopa')
plt.gca().legend(('e^2_parabola','abs_parabola','an_parabola','sin','data'))
plt.show()
```
